- duplicated data

##### Messy Data 
- structual issues 
- Tidy data should have:
       
       Each variable forms a columns 
       Each observation forms a row 
       Each observational unit forms a table

So, we can see from the output above we have a number of different data types.

- assigned_sex is actually a categorical variable. It's ok as a string (object in pandas) but we could change it to categrical to be more specific
- zip code should not be a float, it should be a string. This is because leading zeros may need to exist in the data
- birthdate as a string can sometimes be fine, but really we want some sort of date/time stamp
- state is also down as a string, but again, this could be a 'categorical' variable to be more specific
